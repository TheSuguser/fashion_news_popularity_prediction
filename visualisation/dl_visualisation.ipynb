```python
def switch(probs):
    for prob in probs:
        prob[:,[1,2]] = prob[:,[2,1]]
    return probs
```

```python
def accuracy(preds, y_true):
    '''
    args:
        preds: list of prediction probability
        y_true: one-hot ground truth
    '''
    y_true = np.argmax(y_true, axis=1)
    logits = [np.argmax(pred, axis=1) for pred in preds]
    acc = [np.mean(logit==y_true) for logit in logits]
    
    return acc
```

```python
def recall(preds, y_true):
    y_true = np.argmax(y_true, axis=1)
    logits = [np.argmax(pred, axis=1) for pred in preds]
    recall = [recall_score(y_true, y_pred=logit, average='macro') for logit in logits]
    
    return recall
```

```python
def roc(preds, y_true):
    ra = [roc_auc_score(y_true, y_score=pred) for pred in preds]
    
    return ra
```

```python
def f1(preds, y_true):
    y_true = np.argmax(y_true, axis=1)
    logits = [np.argmax(pred, axis=1) for pred in preds]
    score = [f1_score(y_true, y_pred=logit, average='macro') for logit in logits]
    
    return score
```

```python
import itertools

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.figure()
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    #plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.4f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.savefig('dl_cm_{}.png'.format(title), dpi=300,bbox_inches='tight')
```

```python
def plot_cm(y_true, probs, names=None):
    length = min([p.shape[0] for p in probs])
    #y_true[:, [1,2]] = y_true[:,[2,1]]
    y_true = np.argmax(y_true[-length:],axis=1)
    probs = [p[-length:,:] for p in probs]
    #for p in probs:
     #   p[:,[1,2]] = p[:,[2,1]]
    #probs = [p[:,[1,2]]=p[:,[2,1]] for p in probs]
    preds = [list(np.argmax(p,axis=1)) for p in probs]
    for pred, name in zip(preds,names):
        cm = confusion_matrix(y_true=y_true, y_pred=pred)
        plot_confusion_matrix(cm=cm,
                              classes=['Cold', 'Medium', 'Hot'],
                              normalize=True,
                              title=name)
    #print(len(accs))
    
```

```python
def plot_curve(y_true, preds ,names=None, label=2):
    plt.figure()
    y_true = np.argmax(y_true, axis=1)
    
    for pred in preds:
        fpr, tpr, thr = roc_curve(y_true, pred[:,label], pos_label=label)
        plt.plot(fpr, tpr)
    plt.legend(names)
    
    if label==0:
        plt.title('Positive label: \'COLD\'')
    else:
        plt.title('Positive label: \'HOT\'')        
    
    plt.savefig('dl_roc_{}.png'.format(label), dpi=300,bbox_inches='tight')
    
```
